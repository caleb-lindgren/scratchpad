# PDC data demo

This notebooks demonstrates using the `cptac.pancan` module to stream data tables from the PDC.

First, import the module.

```python
import cptac.pancan as pancan
```

### Available data

The `list_pdc_datasets` function lists the PDC study IDs for all datasets available through `cptac.pancan`. The module groups data by cancer type into classes; thus loading the `PdcBrca` class will provide you the BRCA acetylome, proteome, and phosphoproteome data tables, which come from the listed PDC study IDs.

The quantitative tables each come from a quantDataMatrix query of the PDC database. For each cancer type, we also provide the clinical table from a clinicalPerStudy query.

```python
pancan.list_pdc_datasets()
```

### Loading datasets

To load a dataset, we first download it.

```python
pancan.download("PdcBrca")
```

Then, load the corresponding class.

```python
br = pancan.PdcBrca()
```

To access individual tables, call the class's corresponding "get" method:

```python
br.get_clinical()
```

```python
br.get_proteomics()
```

```python
br.get_phosphoproteomics()
```
