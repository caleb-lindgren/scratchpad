```python
dss = {
    "br": cptac.Brca,
    "cc": cptac.Ccrcc,
    "co": cptac.Colon,
    "en": cptac.Endometrial,
    "gb": cptac.Gbm,
}
```

```python
ads = {
    "hn": cptac.Hnscc,
    "ls": cptac.Lscc,
    "lu": cptac.Luad,
    "ov": cptac.Ovarian
}
```

```python
def load(dss):
    vals = pd.DataFrame()
    for name in dss.keys():
        cna = dss[name]().get_CNV()
        flat = pd.DataFrame({"cna_vals": cna.values.flatten()})
        vals = vals.append(flat)
        
    return vals
```

```python
first = load(dss)
```

```python
sec = load(ads)
```

```python
flat = first.append(sec)
```

```python
filtered = flat.loc[(flat["cna_vals"] > -2) & (flat["cna_vals"] < 2)]
```

```python
alt.Chart(filtered).mark_bar().encode(
    alt.X(
        "cna_vals",
        bin=alt.Bin(step=0.05),
        scale=alt.Scale(domain=(-2, 2))
    ),
    y="count()"
).properties(
    width=2000,
    title="All cancer types"
)
```
