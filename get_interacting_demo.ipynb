# cptac.utils: Get interacting proteins demo

There are functions to get interacting protein lists from four sources: Wikipathways, Bioplex, Biogrid, and STRING.

```python
import cptac.utils as ut
```

### Wikipathways

```python
wp = ut.get_interacting_proteins_wikipathways("EGFR")
print(f"First ten results: {wp[:10]}")
print(f"Total number of results: {len(wp)}")
```

### Bioplex

```python
bp = ut.get_interacting_proteins_bioplex("EGFR")
print(f"First ten results:\n{bp[:10]}")
print(f"Total number of results: {len(bp)}")
```

## Note: `num_results` parameter

The Biogrid and STRING functions both access web APIs that allow you to specify a number of results you want returned. You can change the value using the `num_results` parameter. The default is 25.

### Biogrid

```python
bg = ut.get_interacting_proteins_biogrid("EGFR", num_results=25)
print(f"First ten results:\n{bg[:10]}")
print(f"Total number of results: {len(bg)}")
```

### STRING

```python
st = ut.get_interacting_proteins_string("EGFR", num_results=25)
print(f"First ten results:\n{st[:10]}")
print(f"Total number of results: {len(st)}")
```
