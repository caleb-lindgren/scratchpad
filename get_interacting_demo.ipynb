## Note: `num_results` parameter

The Biogrid and STRING functions both access web APIs that allow you to specify a number of results you want returned. You can change the value using the `num_results` parameter. The default is 25.

```python
ut.get_interacting_proteins_biogrid("EGFR", num_results=25)
```

```python
ut.get_interacting_proteins_string("EGFR", num_results=25)
```

```python
ut.get_interacting_proteins_bioplex("EGFR")
```
