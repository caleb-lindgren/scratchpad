```python
import pandas as pd
import numpy as np
import cptac
import cptac.utils as ut
import operator
from IPython.display import Image
```

```python
en = cptac.Endometrial()
```

```python
tumor = en.get_proteomics(tissue_type="tumor").mean()
normal = en.get_proteomics(tissue_type="normal").mean()

both = tumor.combine(normal, operator.sub)

prot = en.get_proteomics().transpose()
```

```python
prot
```

```python
ut.reactome_pathway_overlay(
    df=prot, 
    pathway="R-HSA-140834", 
    open_browser=True)
```

```python
image_path = ut.reactome_pathway_overlay(
    df=both,
    pathway="R-HSA-877300",
    open_browser=False,
    export_path="test.png")

Image(image_path[1])
```

```python
image_path = ut.reactome_pathway_overlay(
    df=both,
    pathway="R-HSA-8950505",
    open_browser=False,
    export_path="test.png")

Image(image_path[1])
```
