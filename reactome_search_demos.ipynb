# `cptac.utils` Reactome search function demos

This notebook goes over two  functions I've added to `cptac.utils`. They both query the [Reactome Content Service API](https://reactome.org/dev/content-service) for information related to pathway membership.

```python
import cptac.utils as ut
```

## `search_reactome_pathways_with_proteins`

This function takes a single protein or a list of proteins as input, and returns a dataframe telling you all pathways that contain each protein.

```python
df = ut.search_reactome_pathways_with_proteins("BRCA1")
df.head()
```

```python
df = ut.search_reactome_pathways_with_proteins(["PTEN", "PIK3CA"])
df.head()
```

```python
df.tail()
```

If you query for an identifier for which no information is found, a warning is issued. You can silence these by setting the `quiet` parameter to `True`.

```python
df = ut.search_reactome_pathways_with_proteins(["PTEN", "yatahey"])
df.head()
```

```python
df = ut.search_reactome_pathways_with_proteins(["PTEN", "yatahey"], quiet=True)
df.head()
```

# `search_reactome_proteins_in_pathways`

This function takes a single Reactome pathway identifier or a list of them, and tells you all proteins and gene products contained in it or them.

```python
df = ut.search_reactome_proteins_in_pathways("R-HSA-8935690")
df.head()
```

```python
df = ut.search_reactome_proteins_in_pathways(["R-HSA-8935690", "R-HSA-418346"])
df.head()
```

```python
df.tail()
```

Like the previous function, it issues a warning when identifiers are not found, and those warnings can be repressed with the `quiet` parameter.

```python
df = ut.search_reactome_proteins_in_pathways(["R-HSA-8935690", "yatahey"])
df.head()
```

```python
df = ut.search_reactome_proteins_in_pathways(["R-HSA-8935690", "yatahey"], quiet=True)
df.head()
```
