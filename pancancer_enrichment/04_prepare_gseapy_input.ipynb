## Setup

```python
TIME_START = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')

STEP04_DIR = "step04_outputs"

if not os.path.isdir(STEP04_DIR):
    os.mkdir(STEP04_DIR)
```

## Prepare data

```python
# Create a dictionary of the datasets
# We don't load them yet--we'll do it one at a time to save RAM
dss = {
#     "brca": cptac.Brca, # Exclude BRCA--no normal samples
    "ccrcc": cptac.Ccrcc,
    "colon": cptac.Colon,
    "endometrial": cptac.Endometrial,
    "gbm": cptac.Gbm,
    "hnscc": cptac.Hnscc,
    "lscc": cptac.Lscc,
    "luad": cptac.Luad,
    "ovarian": cptac.Ovarian,
}
```

```python
# We define this as a separate function so that its objects will
# go out of scope each time the function returns, which will
# clear up RAM

def prepare_expr_data(cancer_type, datasets, export_path):
    
    ds = datasets[cancer_type]()
    prot = ds.get_proteomics()
    
    if prot.columns.nlevels == 1:
        prot = prot.transpose()
        
    elif prot.columns.nlevels == 2 and "Database_ID" in prot.columns.names:
        prot = ut.reduce_multiindex(prot, levels_to_drop="Database_ID").\
            transpose().\
            groupby("Name").\
            mean()
        
    else:
        raise Exception(cancer_type)
        
    prot.index.name = "NAME"
        
    file_name = f"{cancer_type}_{ds.version()}_prot.tsv"
    file_path = os.path.join(export_path, file_name)
    
    prot.to_csv(file_path, sep='\t')
```

```python
for cancer_type in dss.keys():
    prepare_expr_data(cancer_type, dss, STEP04_DIR)
```
