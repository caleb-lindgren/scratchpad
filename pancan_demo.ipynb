# Pancan data access demo

First, import the package.

```python
import cptac.pancan as pc
```

Authentication through your Box account is required when you download files.

```python
pc.pancan_download("pancanbrca")
```

```python
pc.pancan_download("pancangbm")
```

## Load the BRCA dataset

```python
br = pc.PancanBrca()
```

```python
br.get_clinical(source="mssm")
```

```python
br.get_somatic_mutation(source="washu")
```

```python
br.get_proteomics(source="umich")
```

### PDC data

We can also access data from the PDC.

```python
br.get_proteomics(source="pdc")
```

```python
br.get_clinical(source="pdc")
```
